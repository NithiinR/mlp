```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from math import comb
p = 0.3
q = 1-p
n = 100
fig, (ax1) = plt.subplots()
ax.set(title="binomial distribution",
       xlabel = "k",
       ylabel = "proboability of getting k success in ${n} trials"
      )
X = [i for i in range(0,n)]
Y = [comb(n,x)*(p**x)*(q**(n-x)) for x in X]
ax1.hist(X);
```
