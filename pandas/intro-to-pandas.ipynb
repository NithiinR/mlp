## Manipulating Data 2.0

```python
#adding columns 
#column from series
seats_col = pd.Series([5 for _ in range(0,5)])
# New col called Seats
car_sales["Seats"] = seats_col
car_sales
```

```python
car_sales["Seats"].fillna(5,inplace=True)
car_sales.tail()
```

```python
# Column from python list 
# list has to be same size of dataFrome rows unlike series
fuel_eco = [7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales['Fuel per 100KM'] = fuel_eco
car_sales
```

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales.head()
```

```python
#Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales.head()
```

```python
car_sales["Passed Road Safety"] = True
car_sales.dtypes
```

```python
#To remove a column
car_sales.drop("Total fuel used",axis=1,inplace=True)
car_sales
```
